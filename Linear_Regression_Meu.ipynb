```python
from sklearn.datasets import make_regression 
import matplotlib.pyplot as plt
#gerando  uma massa de dado:
x,y = make_regression(n_samples=200,n_features=1, noise=00)
#mostrando os dados no grafico
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(x,y)     # Faz os calculos para gerara as retas
coef_lin = modelo.intercept_   # Coeficiente linear 
coef_ang = modelo.coef_        # Corficiente angular
coef_ang = coef_ang[0]
```

```python
import numpy as np
plt.scatter(x,y)
xreg = np.arange(-3,3,1)
plt.plot(xreg,coef_ang*xreg-coef_lin,c='red')
plt.show()
```
